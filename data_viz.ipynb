```python
# Ploting mathematical functions
def func(x):
    return 0.5 * np.exp(x) + 1

a, b = 0.5, 1.5 #integral limit
x = np.linspace(0,2) 
y = func(x)
Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b,0)] #coordinates of polygon to be plotted
```

```python
from matplotlib.patches import Polygon
fig, ax = plt.subplots(figsize=(10,6))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(bottom=0)
poly = Polygon(verts, facecolor='0.5', edgecolor='0.1')
ax.add_patch(poly)
plt.text(0.5*(a+b), 1, r'$\int_a^b f(x)\mathrm{d}x$', horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
ax.set_xticks((a,b))
ax.set_xticklabels(('$a$', '$b$'))
ax.axhline(func(a), c='r')
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
plt.show()
```

## Static 3D Plotting

```python
strike = np.linspace(50, 150, 24)
```

```python
ttm = np.linspace(0.5, 2.5, 24)
strike, ttm = np.meshgrid(strike, ttm)
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12, 8))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2, 
                      cmap=plt.cm.coolwarm, linewidth=0.5,
                      antialiased=True)
ax.set_xlabel('strike')
ax.set_ylabel('time_to_maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

```python
# Changing marker and viewing angle
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60)
ax.scatter(strike, ttm, iv, zdir='z', s=25, c='b', marker='^')
ax.set_xlabel('strike')
ax.set_ylabel('time_to_maturity')
ax.set_zlabel('implied_vol')
plt.show()
```

```python
# Offline version of Plotly supported by Cufflunks integration with Pandas
```

```python
import pandas as pd
import cufflinks as cf
import plotly.offline as plyo

plyo.init_notebook_mode(connected=True)
```

```python
a = np.random.standard_normal((250, 5)).cumsum(axis=0)
index = pd.date_range('2019-1-1', freq='B', periods=len(a))
df = pd.DataFrame(100 + 5 * a, columns=list('abcde'), index=index)
df.head()
```

```python
plyo.iplot(df.iplot(asFigure=True, theme='polar'), 
                   #image = 'png', 
                   filename='ply_01')
```

```python
plyo.iplot(df[['a', 'b']].iplot(asFigure=True,
          theme='polar', 
          title='A Time Series Plot', 
          xTitle='date',
          yTitle='value',
          mode={'a': 'markers', 'b': 'lines+markers'},
          symbol={'a': 'circle', 'b': 'diamond'},
          size=3.5,
          colors={'a': 'blue', 'b': 'magenta'}),
          #image='png',
          filename='ply_02')
```

```python
plyo.iplot(
    df.iplot(kind='hist',
            subplots=True, 
            bins=15, 
            asFigure=True),
    #image='png',
    filename='ply_03')
```

```python
# data from FXCM Forex Capital Markets Ltd.
raw = pd.read_csv('data/EUR_USD Historical Data.csv',
                 index_col=0, parse_dates=True) 
```

```python
quotes = raw[['Open', 'High', 'Low', 'Price']]
quotes['Close'] = quotes.Price
quotes = quotes.iloc[-60:]
quotes.tail()
```

```python
qf = cf.QuantFig(quotes, 
                title='EUR/USD Exchange Rate',
                legend='top',
                name='EUR/USD')

plyo.iplot(qf.iplot(asFigure=True),
          #image='png',
          filename='qf_01')
```
