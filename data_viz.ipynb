```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
%matplotlib inline 
```

```python
import numpy as np
np.random.seed(1000)
```

## Static 2D Plotting

```python
y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(y.cumsum())
plt.show()
```

```python
plt.plot(y.cumsum())
#plt.grid(False)
plt.axis('equal')
plt.show()
```

```python
plt.plot(y.cumsum())
#plt.grid(False)
plt.axis('scaled')
plt.show()
```

```python
plt.plot(y.cumsum())
#plt.grid(False)
plt.axis('equal')
plt.show()
```

```python
plt.plot(y.cumsum())
plt.axis('tight')
plt.show()
```

```python
plt.plot(y.cumsum())
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1, np.max(y.cumsum())+1)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(y.cumsum(), color = 'k', lw=1.5)
plt.plot(y.cumsum(),  'D', color='r')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Agly Chart Title')
plt.show()
```
