```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
%matplotlib inline 
```

```python
import numpy as np
np.random.seed(1000)
```

## Static 2D Plotting

```python
y = np.random.standard_normal(20)
x = np.arange(len(y))
plt.plot(x,y)
plt.show()
```

```python
plt.plot(y)
plt.show()
```

```python
plt.plot(y.cumsum())
plt.show()
```

```python
plt.plot(y.cumsum())
#plt.grid(False)
plt.axis('equal')
plt.show()
```

```python
plt.plot(y.cumsum())
#plt.grid(False)
plt.axis('scaled')
plt.show()
```

```python
plt.plot(y.cumsum())
#plt.grid(False)
plt.axis('equal')
plt.show()
```

```python
plt.plot(y.cumsum())
plt.axis('tight')
plt.show()
```

```python
plt.plot(y.cumsum())
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1, np.max(y.cumsum())+1)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(y.cumsum(), color = 'k', lw=1.5)
plt.plot(y.cumsum(),  'D', color='r')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Agly Chart Title')
plt.show()
```

```python
y = np.random.standard_normal((20,2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10,6))
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,0], 'D', color='r')
plt.plot(y[:,1], lw=1.5, label='2nd')
plt.plot(y[:,1], 'D', color='k')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Much Better Plot')
plt.show()
```

```python
#increasing the scale
y[:,0] = y[:,0]*100
```

```python
fig, ax1 = plt.subplots()
plt.plot(y[:,0], 'b', lw=1.5, label='1st')
plt.plot(y[:,0], 'ro')
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')

ax2 = ax1.twinx()
plt.plot(y[:,1], 'g', lw=1.5, label='2nd')
plt.plot(y[:,1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')

plt.show()
```
